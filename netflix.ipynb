```python
# Load the dataset
file_path = 'netflix_content_2023.csv'  # Replace with your file name or path
data = pd.read_csv(file_path)

# Convert 'Hours Viewed' column to numeric (remove commas and handle strings)
data['Hours Viewed'] = data['Hours Viewed'].str.replace(',', '').astype(float)

# Count the number of titles by language
language_distribution = data['Language Indicator'].value_counts()

# Plot language distribution
plt.figure(figsize=(10, 5))
language_distribution.plot(kind='bar', title='Distribution of Content by Language', color='skyblue')
plt.xlabel('Language')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()

# Group by language and sum hours viewed
language_interest = data.groupby('Language Indicator')['Hours Viewed'].sum()

# Plot the resultsplt.figure(figsize=(10, 5))
language_interest.sort_values(ascending=False).plot(kind='bar', title='Audience Interest by Language', color='green')
plt.xlabel('Language')
plt.ylabel('Total Hours Viewed (in billions)')

plt.show()
```

```python
# Count global availability
global_availability = data['Available Globally?'].value_counts()

# Plot the results
global_availability.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6), title='Global Availability of Content')
plt.ylabel('')
plt.show()

# Create a boxplot to compare Hours Viewed for globally available vs. region-specific content
plt.figure(figsize=(8, 5))
sns.boxplot(x='Available Globally?', y='Hours Viewed', data=data, palette='Set2')
plt.title('Viewership by Global Availability')
plt.xlabel('Available Globally')
plt.ylabel('Hours Viewed')
plt.show()
```

```python
# Group by content type and sum hours viewed
content_viewership = data.groupby('Content Type')['Hours Viewed'].sum()

# Plot viewership by content type
content_viewership.plot(kind='bar', title='Total Hours Viewed by Content Type', color='orange', figsize=(8, 5))
plt.xlabel('Content Type')
plt.ylabel('Total Hours Viewed')
plt.show()
```

```python
# Sort by Hours Viewed to find the most-watched titles
top_titles = data[['Title', 'Hours Viewed']].sort_values(by='Hours Viewed', ascending=False).head(10)

# Display the top titles
print("Top 10 Most-Watched Titles:")
print(top_titles)

# Bar plot of the top 10 titles
plt.figure(figsize=(12, 6))
sns.barplot(x='Hours Viewed', y='Title', data=top_titles, palette='viridis')
plt.title('Top 10 Most-Watched Netflix Titles')
plt.xlabel('Hours Viewed')
plt.ylabel('Title')
plt.show()
```

```python
# Group by language and sum hours viewed
language_interest = data.groupby('Language Indicator')['Hours Viewed'].sum()

# Plot the results
plt.figure(figsize=(10, 5))
language_interest.sort_values(ascending=False).plot(kind='bar', title='Audience Interest by Language', color='pink')
plt.xlabel('Language')
plt.ylabel('Total Hours Viewed')
plt.show()
```
